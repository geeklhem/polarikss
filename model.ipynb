# Parameters

These are the physical measured physical constants

```python
param = {}
paramA = {}
#Discretisation
param["L"] = 134.6 # µm 
param["T"] = 700 #s 

# Kinetics
param["kon"] = 4e-2 # μm s−1
param["koff"] = 7e-3 # s-1
paramA["kon"] = 8.58e-3 # µm s-1
paramA["koff"] = 5.4e-3 # s-1
param["psi"] = 0.174 #µm-1

# Concentration
paramA["rho"] = 1.56 #µm-3
param["rho"] = 1 #µm--3

# Cell volume
paramA["V"] = 1 #µm3


# Diffusion
param["D"] = 0.15 #µm^2.s-1
paramA["D"] = 0.28 #µm^2.s-1

# Flow 
param["F"] = pandas.read_table("flow-myosin-um-per-min.txt",header=None).values
param["xi"] = 20
paramA["xi"] = 0.01

# Copy the parameter of P in A
for k,v in param.items():
    if k not in paramA:
        paramA[k] = v
```

# Model definition

The model definition we went for a foward euler discretisation of the EDP.

In the following you have the simplest implementaton of the model, using only diffusion.

```python
def diffusion(p,param):
    """ The simplest form of the model: Diffusion only
    
    Args:
        p (np.array): A timexSpace array, the first line give
            the initial condtion, its shape gives the number of
            discretization volumes.             
        param (dict): Parameters
        
    Returns:
        (np.array): with a timestep by line. 
    
    """
    dt = param["T"]/float(p.shape[0]) #Time discretisation.
    l = param["L"]/float(p.shape[1])  #Space discretisation.

    for t in range(p.shape[0]-1):
        p[t+1,:] = (p[t,:]+
                    dt*(  param["D"] * l**-2 * (np.concatenate((p[t,1:],[p[t,0]]))   -p[t,:])
                        + param["D"]  * l**-2 * (np.concatenate(([p[t,-1]],p[t,:-1])) -p[t,:])
                    )) 
    return p

        
def display(p,cyto=None): 
    """Display function
    
    Args:
        p (np.array): A timeXspace array."""
    plt.subplot(1,2,1)

    plt.imshow(p,interpolation="nearest",aspect="auto",cmap="rainbow")
    plt.xlabel("distance (um)")
    plt.ylabel("time (sec)")
    plt.subplot(1,2,2)
    
    plt.plot(p[-1,:],label="t=final")
    plt.plot(p[0,:],label="t=0")
    plt.plot(p[1,:],label="t=1")
    #plt.plot(p[2,:],label="t=2")
    
    plt.legend()
    plt.show()
    
    if cyto is not None:
        plt.plot(cyto)
        
    print "t=0", p[0,:5]
    print "t=1", p[1,:5]
    print "t=f", p[-1,:5]
```

# Convergence plot

Convergence plots are of critical importance in ordre to assess the correctness of our discretization scheme and choose the time and space step size.

The idea is to consider a simple case where an analytical solution is known (for instance the diffusion of a dirac 
distribution gives a gaussian) and compare it to the result given by our model for different resolution.

```python
# Convergence in time...

#Analytical solution A(x,t) = 1/np.sqrt(4 \pi D t) * exp(-x**2 / 4Dt )
param["T"] = 100 #s
plt.rcParams['figure.figsize'] =  15,7
analytical = [(
               (4*np.pi*param["D"]*param["T"])**-1/2
                * np.exp((-(-50+x)**2)*(4*param["D"]*param["T"])**-1)
               ) for x in range(100)]

plt.subplot(1,2,1)
plt.plot(range(100),analytical,'-.',label="Analytical solution",)
        
e = [] #error
N = [] #number of steps

for n in np.arange(0,3.2,0.1):
    p = np.zeros((10**n, 100))
    p[0,int(p.shape[1]/2)] = 1 #A dirac
    p = diffusion(p,param)
    
    # add
    N.append(p.shape[0])
    e.append(np.max(abs(p[-1,:]-analytical)))
    if N[-1]%10 == 0:
        plt.plot(range(100),p[-1,:],alpha=0.5,label="N={}".format(N[-1]))
        
plt.ylim(0,0.1)
plt.legend()
plt.subplot(1,2,2)
plt.plot(N,e)
plt.xlabel("Number of timestep")
plt.ylabel("Maximal Error")
#plt.yscale('log')
plt.xscale('log')
```

# Adding the cytosol

```python
def cytosol(p,param):
    """ Run the model with a forward euler discretization.
    
    Args:
        p (np.array): A timexSpace array, the first line give
            the initial condtion, its shape gives the number of
            discretization volumes.             
        param (dict): Parameters
        
    Returns:
        (np.array): with a timestep by line. 
    
    """
    dt = param["T"]/float(p.shape[0]) #Time discretisation.
    l = param["L"]/float(p.shape[1])  #Space discretisation.
    cyto = np.zeros(p.shape[0])
    cyto[0] = param["rho"]
    
    # Time loop
    for t in range(p.shape[0]-1):
        p[t+1,:] = (p[t,:]+
                    dt*(cyto*param["kon"]*param["psi"] 
                        - p[t,:]*param["koff"] 
                        + param["D"] * l**-2 * (np.concatenate((p[t,1:],[p[t,0]]))   -p[t,:])
                        + param["D"] * l**-2 * (np.concatenate(([p[t,-1]],p[t,:-1])) -p[t,:])
                    )) 
        cyto[t+1] = cyto[t] + dt*(-p.shape[1]*cyto*param["kon"]*param["psi"] + np.sum(p[t,:])*param["koff"])
    return p,cyto
```

```python
# First we definie an empy matrix (time*space) that we will fill step by step.
p = np.zeros((3000, 65))
# Initial conditions : a dirac in the middle.
p[0,int(p.shape[1]/2)] = 10000
p[0,5] = 10000
p[0,25] = 10000


# Run the simulation ! 
p,cyto = cytosol(p,param)
display(p,cyto)
```
